```python
def J_cav(g,H_free, Trans_12, Trans_13, Trans_23, a, nh,nf,nc,h,kb,gamma_h,gamma_c,kappa,omega_d,proj_2,f,omega_f,omega_2):
    
    gamma_1=(nh+1)*gamma_h #### unsicher wegen vorfaktor 1/2 
    gamma_2=(nh)*gamma_h
    gamma_3=(nc+1)*gamma_c
    gamma_4=(nc)*gamma_c
    kappa_52=(nf+1)*kappa/2 ####goes to zero
    kappa_6=(nf)*kappa/2

    A1=Trans_13
    A2=Trans_13.dag()
    A3=Trans_23
    A4=Trans_23.dag()
    A5=a
    A6=a.dag()
    
    c_op_list=[]

    c_op_list.append(np.sqrt(gamma_1)*A1)
    c_op_list.append(np.sqrt(gamma_2)*A2)
    c_op_list.append(np.sqrt(gamma_3)*A3)
    c_op_list.append(np.sqrt(gamma_4)*A4)
    c_op_list.append(np.sqrt(kappa_5)*A5)
    c_op_list.append(np.sqrt(kappa_6)*A6)

    
    
    
    V=f*a.dag()+f*a
    H_int=h*g*(Trans_12*a.dag()+a*Trans_12.dag())
        
        
    #H_free=omega_1*proj_1+h*omega_2*proj_2+h*omega_3*proj_3+h*omega_f*a.dag()*a 
    Hdilde=H_int+V +(omega_2-(30+omega_d))*(a.dag()*a)+(omega_f-omega_d)*proj_2  
    rho=steadystate(Hdilde,c_op_list)
    
    
    
    J_cav=np.trace(H_free*(c_op_list[4]*rho*c_op_list[4].dag()-1/2*c_op_list[4].dag()*c_op_list[4]*rho-1/2*rho*c_op_list[4].dag()*c_op_list[4]+c_op_list[5]*rho*c_op_list[5].dag()-1/2*c_op_list[5].dag()*c_op_list[5]*rho-1/2*rho*c_op_list[5].dag()*c_op_list[5]))
    
    return J_cav
```
